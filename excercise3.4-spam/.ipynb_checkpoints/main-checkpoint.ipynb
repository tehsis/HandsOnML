```python
import os

def read_from_dirs(dirs):
    results = []
    for dir in dirs:
        files_paths = os.listdir(dir)
        for fp in files_paths:
            results.append(get_file_content(os.path.join(dir, fp)))
    return results

ham = read_from_dirs(["easy_ham", "easy_ham_2", "hard_ham"])
spam = read_from_dirs(["spam", "spam_2"])
```

```python
from collections import Counter
import re

def word_count(str):
    words = re.findall(r'\b\w+\b', str.lower())
    word_counts = Counter(words)
    return word_counts
```
